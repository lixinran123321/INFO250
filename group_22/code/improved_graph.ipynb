# Improved Graph

```python
# Import library
import matplotlib.pyplot as plt

# Set font
plt.rc('font',family='Times New Roman') 

# Construct coordinate
X = [str(i) for i in range(1970, 2031, 10)]
Y = [2.9, 4.3, 6.8, 8.8, 9.9, 14.3, 19.7]
x = [i for i in range(0,7)]

# Draw image
f,ax = plt.subplots(dpi=200)
plt.plot(X, Y, markersize=5,marker = 'o', label="haha")
plt.ylim(bottom=0.)
plt.xlabel("Year")
plt.text(-2,21.5,"Population over 65(%)", fontsize=10)
plt.title("Trend of aging population in Taiwan", fontsize=15, pad=30)
for x1, y1 in zip(x, Y):
    if x1==0 or x1==6:
        plt.text(x1,y1+0.35, str(y1), ha='center', va='bottom', fontsize=12)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()

# Save it as a svg vector image.
plt.savefig("improved_image.svg")
```
